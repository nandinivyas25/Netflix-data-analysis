```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['type'])
```

```python
df['type'].value_counts().plot(kind='bar',color='pink')
plt.title('count of movies vs tv shows')
plt.xlabel('type')
plt.ylabel('count')
plt.grid(True)
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', color='blue')
plt.title('Top 10 countries by netflix content')
plt.xlabel('Number of titles')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
df['release_year'].value_counts().sort_index().plot(kind='bar', color='gold')
plt.title('Titles Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
genres = df['listed_in'].str.split(', ', expand=True).stack()
genres.value_counts().head(10).plot(kind='bar', color='purple')
plt.title('Top 10 Most Common Genres')
plt.ylabel('Frequency')
plt.show()
```
